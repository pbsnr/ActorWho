### Netflix film name

```python
import requests
from bs4 import BeautifulSoup
```

```python
page = requests.get("https://www.netflix.com/watch/70250213?trackId=14170286&tctx=7%2C1%2Cf19da7c2-b4c6-4d38-8462-535983d4c67a-90078927%2C7be0cb14-ab37-4222-88b5-5e7e754ccb76_71759217X3XX1611081880258%2C7be0cb14-ab37-4222-88b5-5e7e754ccb76_ROOT%2C")
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
soup
```

```python
movies = soup.findAll('title')
title = str(movies)
```

```python
title = title.replace('[<title>','')
title = title.replace('| Netflix</title>]','')
```

### FINAL SCRIPT

```python
import requests
from bs4 import BeautifulSoup
link = ''
if 'netflix.com' in link:
    page = requests.get(link)
    soup = BeautifulSoup(page.content, 'html.parser')
    movies = soup.findAll('title')
    title = str(movies)
    title = title.replace('[<title>','')
    title = title.replace('| Netflix</title>]','')
    print('Movie title is: ',title)
if 'primevideo.com' in link: 
    page = requests.get(link)
    soup = BeautifulSoup(page.content, 'html.parser')
    movies = soup.findAll('title')
    title = str(movies[0])
    title = title.replace('<title>Prime Video:','')
    title = title.replace('</title>','')
    print('Movie title is: ',title)
else:
    print('Application requires Netflix or PrimeVideo only')
```
